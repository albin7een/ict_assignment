```python


def input_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                print("Amount cannot be negative.")
                continue
            return value
        except:
            print("Enter a valid amount.")


def input_date():
    while True:
        date = input("Enter date (YYYY-MM-DD): ")
        parts = date.split("-")
        if len(parts) == 3 and len(parts[0]) == 4 and len(parts[1]) == 2 and len(parts[2]) == 2:
            return date
        else:
            print("Invalid date format.")


# ---------------- Core Functions ----------------

def set_budget():
    global monthly_budget
    monthly_budget = input_float("Enter monthly budget amount: ")
    print("Monthly budget set successfully.")


def add_expense():
    date = input_date()
    category = input("Enter expense category: ").strip()
    amount = input_float("Enter expense amount: ")

    expenses.append({
        "date": date,
        "category": category,
        "amount": amount
    })

    print("Expense added successfully.")


def monthly_summary():
    month = input("Enter month (YYYY-MM): ")
    total = 0

    for exp in expenses:
        if exp["date"].startswith(month):
            total += exp["amount"]

    print("Total expenses for", month, ":", total)

    if monthly_budget > 0 and total > monthly_budget:
        print("âš  Budget Breach Alert! You exceeded your budget.")


def category_report():
    category_total = {}

    for exp in expenses:
        cat = exp["category"]
        category_total[cat] = category_total.get(cat, 0) + exp["amount"]

    print("\n--- Category-wise Spending ---")
    for cat, amt in category_total.items():
        print(cat, ":", amt)


def analytical_insights():
    if not expenses:
        print("No expenses to analyze.")
        return

    day_total = {}
    category_total = {}

    for exp in expenses:
        date = exp["date"]
        category = exp["category"]
        amount = exp["amount"]

        day_total[date] = day_total.get(date, 0) + amount
        category_total[category] = category_total.get(category, 0) + amount

    highest_day = ""
    highest_amount = 0

    for day in day_total:
        if day_total[day] > highest_amount:
            highest_amount = day_total[day]
            highest_day = day

    dominant_category = ""
    max_category_amount = 0

    for cat in category_total:
        if category_total[cat] > max_category_amount:
            max_category_amount = category_total[cat]
            dominant_category = cat

    print("\n--- Analytical Insights ---")
    print("Highest Spending Day:", highest_day, "Amount:", highest_amount)
    print("Dominant Spending Category:", dominant_category)
    print("Amount Spent:", max_category_amount)


def display_expenses():
    if not expenses:
        print("No expenses recorded.")
        return

    print("\nDate | Category | Amount")
    for exp in expenses:
        print(exp["date"], "|", exp["category"], "|", exp["amount"])


# ---------------- Main Menu ----------------

def main_menu():
    while True:
        print("\n===== Personal Expense Tracker =====")
        print("1. Set Monthly Budget")
        print("2. Add Expense")
        print("3. Monthly Summary")
        print("4. Category-wise Report")
        print("5. Analytical Insights")
        print("6. Display All Expenses")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            set_budget()
        elif choice == "2":
            add_expense()
        elif choice == "3":
            monthly_summary()
        elif choice == "4":
            category_report()
        elif choice == "5":
            analytical_insights()
        elif choice == "6":
            display_expenses()
        elif choice == "7":
            print("Exiting Expense Tracker. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")


# ---------------- Run Program ----------------

main_menu()
```
